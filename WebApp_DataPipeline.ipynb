# FLASK API to display charts on Web App

```python
from flask import Flask, render_template
import pandas as pd
import json
import plotly
import plotly.express as px

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/chart1')
def chart1():
    # Bar Chart displaying Top 10 Countries With highest population in 2007.
    fig = px.bar(df1, x='country', y='pop', color='country',title='Top 10 Countries With Highest Population in 2007')

    graphJSON = json.dumps(fig, cls=plotly.utils.PlotlyJSONEncoder)
    return render_template('graphs.html', graphJSON=graphJSON)

@app.route('/chart2')
def chart2():
    # Bar Chart displaying Top 10 Countries With highest population in 1952.
    fig = px.bar(df2, x='country', y='pop', color='country',title='Top 10 Countries With Highest Population in 1952')

    graphJSON = json.dumps(fig, cls=plotly.utils.PlotlyJSONEncoder)
    return render_template('graphs.html', graphJSON=graphJSON)

@app.route('/chart3')
def chart3():
    # Scatter Chart displaying Top 5 Countries With Highest Life Expectancy in 1952.
    fig = px.scatter(df3, x="country", y="lifeExp", color='country', size='lifeExp',title='Top 5 Countries With Highest Life Expectancy in 1952')

    graphJSON = json.dumps(fig, cls=plotly.utils.PlotlyJSONEncoder)
    return render_template('graphs.html', graphJSON=graphJSON)

@app.route('/chart4')
def chart4():
    # Scatter Chart displaying Top 5 Countries With Highest Life Expectancy in 2007.
    fig = px.scatter(df4, x="country", y="lifeExp", color='country', size='lifeExp',title='Top 5 Countries With Highest Life Expectancy in 2007')

    graphJSON = json.dumps(fig, cls=plotly.utils.PlotlyJSONEncoder)
    return render_template('graphs.html', graphJSON=graphJSON)

@app.route('/chart5')
def chart5():
    # Line Chart displaying GDP Percap of Selected Countries between Years 1952 to 2007.
    fig = px.line(df5, x="year", y="gdpPercap", color='country',title='GDP Percap of Selected Countries between Years 1952 to 2007')

    graphJSON = json.dumps(fig, cls=plotly.utils.PlotlyJSONEncoder)
    return render_template('graphs.html', graphJSON=graphJSON)


@app.route('/chart6')
def chart6():
    # Line Chart displaying Life Expectancy in Selected Countries between Years 1952 to 2007.
    fig = px.line(df6, x="year", y="lifeExp", color='country',title='Life Expectancy in Selected Countries between Years 1952 to 2007')

    graphJSON = json.dumps(fig, cls=plotly.utils.PlotlyJSONEncoder)
    return render_template('graphs.html', graphJSON=graphJSON)

if __name__ == "__main__":
    app.run()
```
